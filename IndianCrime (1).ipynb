```python
features = ['Report Number', 'Date Reported', 'Date of Occurrence',
       'Time of Occurrence', 'City', 'Crime Code', 'Crime Description',
       'Victim Age', 'Victim Gender', 'Weapon Used', 'Crime Domain',
       'Police Deployed', 'Case Closed', 'Date Case Closed']
for feature in features:
    null_check = pd.isnull(df[feature])
    print(df[null_check])
```

```python
df ['City'].unique()
```

```python
df_city = df.groupby('City').size().reset_index(name = 'Numbers')
df_city
```

```python
plt.figure(figsize = (20, 5))
plt.xlabel('City')
plt.title('Arrests In City')
plt.ylabel('Numbers')
x = df_city['City']
y = df_city['Numbers']
plt.bar(x, y)
plt.grid(which = 'both', linestyle = ':')
plt.savefig('City.jpg', dpi = 480)
```

```python
q_1 = df['Victim Age'].quantile(0.25)
q_3 = df['Victim Age'].quantile(0.75)
iqr = q_3 - q_1
lower = q_1 - 1.5 * iqr
upper = q_3 + 1.5 * iqr
```

```python
upper_array = np.where(df['Victim Age'] >= upper)[0]
lower_array = np.where(df['Victim Age'] <= lower)[0]
```

```python
df.drop(index = upper_array, inplace = True)
df.drop(index = lower_array, inplace = True)
```
